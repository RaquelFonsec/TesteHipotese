```python
import scipy.stats as stats
!pip install scipy


# Valores conhecidos
media_populacional = 7.85
desvio_padrao_populacional = 3.32
n_amostras = 45
media_amostral = 11.52
nivel_significancia = 0.05

# Teste de Hipótese
t_statistic = (media_amostral - media_populacional) / (desvio_padrao_populacional / (n_amostras ** 0.5))

# Cálculo do valor crítico de t e do p-valor
graus_de_liberdade = n_amostras - 1
t_critico = stats.t.ppf(1 - nivel_significancia/2, graus_de_liberdade)
p_valor = stats.t.sf(abs(t_statistic), graus_de_liberdade) * 2

# Intervalo de Confiança
erro_padrao = desvio_padrao_populacional / (n_amostras ** 0.5)
intervalo_confianca = stats.norm.interval(1 - nivel_significancia, loc=media_amostral, scale=erro_padrao)

# Resultados
print("Teste de Hipótese:")
print(f"Valor do teste (t): {t_statistic}")
print(f"Valor crítico de t: {t_critico}")
print(f"p-valor: {p_valor}")

print("\nIntervalo de Confiança:")
print(f"Intervalo de confiança: {intervalo_confianca}")
```
